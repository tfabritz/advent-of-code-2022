```python
with open("input.txt") as f:
    lines = f.readlines()
lines = [l.strip() for l in lines]
lines[:5]
```

```python
elves = []
for i in range(0, len(lines)):
    for j in range(0, len(lines[0])):
        if lines[i][j] == "#":
            elves.append((i, j))
# elves
```

```python
def print_elves(elves):
    min_x = 9999999
    min_y = 9999999
    max_x = -1
    max_y = -1

    for e in elves:
        min_x = min(e[0], min_x)
        min_y = min(e[1], min_y)
        max_x = max(e[0], max_x)
        max_y = max(e[1], max_y)

    elf_set = set(elves)
    for i in range(min_x, max_x + 1):
        for j in range(min_y, max_y + 1):
            if (i, j) in elf_set:
                print("#", end="")
            else:
                print(".", end="")
        print()


def count_empty(elves):
    min_x = 9999999
    min_y = 9999999
    max_x = -1
    max_y = -1

    for e in elves:
        min_x = min(e[0], min_x)
        min_y = min(e[1], min_y)
        max_x = max(e[0], max_x)
        max_y = max(e[1], max_y)

    empty_cnt = 0
    elf_set = set(elves)
    for i in range(min_x, max_x + 1):
        for j in range(min_y, max_y + 1):
            if (i, j) not in elf_set:
                empty_cnt += 1
    return empty_cnt


def mv_pos(pos, dir):
    return (pos[0] + dir[0], pos[1] + dir[1])


def has_neighbour(elf, elf_set):
    for i in range(-1, 2):
        for j in range(-1, 2):
            if (i, j) != (0, 0) and mv_pos(elf, (i, j)) in elf_set:
                return True
    return False
```

```python
check_map = {
    "n": [(-1, -1), (-1, 0), (-1, 1)],
    "s": [(1, -1), (1, 0), (1, 1)],
    "w": [(-1, -1), (0, -1), (1, -1)],
    "e": [(-1, 1), (0, 1), (1, 1)],
}


def move_turn(elves, directions):
    elves_res = []
    elf_set = set(elves)
    # dict with target and current elf position who wants to move there
    potential_targets = {}
    # dict with elves and where they want to go
    elf_moves = {}
    for elf in elves:
        proposed_move = (0, 0)

        if not has_neighbour(elf, elf_set):
            elf_moves[elf] = (0, 0)
            continue

        for dir in directions:
            obstructed = False
            for mv in check_map[dir]:
                potential_pos = mv_pos(elf, mv)
                if potential_pos in elf_set:
                    obstructed = True
            if not obstructed:
                proposed_move = check_map[dir][1]
                break

        new_pos = mv_pos(elf, proposed_move)
        if new_pos in potential_targets:
            elf_moves[potential_targets[new_pos]] = (0, 0)
            elf_moves[elf] = (0, 0)
        else:
            potential_targets[new_pos] = elf
            elf_moves[elf] = proposed_move

    moves = 0
    for elf, mv in elf_moves.items():
        if mv != (0, 0):
            moves += 1
        elves_res.append(mv_pos(elf, mv))

    return elves_res, moves


moving_elves = elves.copy()
directions = "nswe"
for i in range(0, 10):
    moving_elves = move_turn(moving_elves, directions)[0]
    directions = directions[1:] + directions[0]

# print_elves(moving_elves)
count_empty(moving_elves)
```
